```python
nidata = pd.read_csv('/content/drive/MyDrive/SemEval Task3/Nikam_emotion_dataset.csv' )   #Nikam_emotion_dataset.csv
nidata.drop(["Unnamed: 0", "Text"], axis=1, inplace = True)
nidata.rename(columns = {"Clean_Text":"text", "Emotion":"emotion"}, inplace=True)
nidata.reindex(columns=["text", "emotion"])
nidata.dropna(inplace=True)
nidata.drop(nidata.index[nidata.emotion=="shame"], inplace=True)
nidata.set_axis([i for i in range(0, len(nidata))], axis = 0, inplace = True)
```

```python
nidata   #neutral:1811   surprise:4062   anger:4286   sadness:6719  joy:11037  disgust:856  fear:5409
```

```python
nidata.rename(columns = {"emotion":"emotions", "text":"utterances_English"}, inplace = True)
```

```python
nidata
```

```python
df_train_val.drop(columns = ["episode","utterances", "speakers"], inplace = True)
```

```python
df_train_val_ni_neutral = pd.concat([df_train_val[df_train_val.emotions=="neutral"], nidata[nidata.emotions=="neutral"]], axis = 0)
df_train_val_ni_contempt = pd.concat([df_train_val[df_train_val.emotions=="contempt"], nidata[nidata.emotions=="contempt"]], axis = 0)
df_train_val_ni_surprise = pd.concat([df_train_val[df_train_val.emotions=="surprise"], nidata[nidata.emotions=="surprise"]], axis = 0)
df_train_val_ni_anger = pd.concat([df_train_val[df_train_val.emotions=="anger"], nidata[nidata.emotions=="anger"]], axis = 0)
df_train_val_ni_sadness = pd.concat([df_train_val[df_train_val.emotions=="sadness"],nidata[nidata.emotions=="sadness"]], axis = 0)
df_train_val_ni_joy = pd.concat([df_train_val[df_train_val.emotions=="joy"], nidata[nidata.emotions=="joy"]], axis = 0)
df_train_val_ni_disgust =  pd.concat([df_train_val[df_train_val.emotions=="disgust"],  nidata[nidata.emotions=="disgust"]], axis = 0)
df_train_val_ni_fear = pd.concat([df_train_val[df_train_val.emotions=="fear"], nidata[nidata.emotions=="fear"]], axis = 0)
```

```python
df_train_val_ni_neutral.set_axis([i for i in range(0, len(df_train_val_ni_neutral))], axis = 0, inplace = True)   #370 because previous notebooks had 370 emotion size for each emotion
df_train_val_ni1000neutral = df_train_val_ni_neutral[df_train_val_ni_neutral.index <= 3000]

df_train_val_ni_contempt.set_axis([i for i in range(0, len(df_train_val_ni_contempt))], axis = 0, inplace = True)
df_train_val_ni1000contempt = df_train_val_ni_contempt[df_train_val_ni_contempt.index <= 3000]

df_train_val_ni_surprise.set_axis([i for i in range(0, len(df_train_val_ni_surprise))], axis = 0, inplace = True)
df_train_val_ni1000surprise = df_train_val_ni_surprise[df_train_val_ni_surprise.index <= 3000]

df_train_val_ni_anger.set_axis([i for i in range(0, len(df_train_val_ni_anger))], axis = 0, inplace = True)
df_train_val_ni1000anger = df_train_val_ni_anger[df_train_val_ni_anger.index <= 3000]

df_train_val_ni_sadness.set_axis([i for i in range(0, len(df_train_val_ni_sadness))], axis = 0, inplace = True)
df_train_val_ni1000sadness = df_train_val_ni_sadness[df_train_val_ni_sadness.index <= 3000]

df_train_val_ni_joy.set_axis([i for i in range(0, len(df_train_val_ni_joy))], axis = 0, inplace = True)
df_train_val_ni1000joy = df_train_val_ni_joy[df_train_val_ni_joy.index <= 3000]

df_train_val_ni_disgust.set_axis([i for i in range(0, len(df_train_val_ni_disgust))], axis = 0, inplace = True)
df_train_val_ni1000disgust = df_train_val_ni_disgust[df_train_val_ni_disgust.index <= 3000]

df_train_val_ni_fear.set_axis([i for i in range(0, len(df_train_val_ni_fear))], axis = 0, inplace = True)
df_train_val_ni1000fear = df_train_val_ni_fear[df_train_val_ni_fear.index <= 3000]
```

```python
df_train_val_ni1000 = pd.concat([df_train_val_ni1000neutral, df_train_val_ni1000contempt, df_train_val_ni1000surprise, df_train_val_ni1000anger, df_train_val_ni1000sadness, df_train_val_ni1000joy, df_train_val_ni1000disgust, df_train_val_ni1000fear], axis = 0)
```
